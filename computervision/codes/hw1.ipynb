```python
from skimage import io
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
from skimage import util
import numpy as np
```

```python
image_rgb = io.imread('99046aab-a2ae-420f-b837-dec74c8c0636.jpeg')
image_gray = rgb2gray(image_rgb)
image_gray_float32 = util.img_as_float32(image_gray)
```

```python
plt.imshow(image_gray_float32,cmap=plt.cm.gray)
```

```python
top_left_3x5 = image_gray_float32[:5, :3]
print("Top-left 3x5 (x by y) block:\n", top_left_3x5)

x, y = 1, 2
pixel_val = image_gray_float32[y-1, x-1]  
print(f"Pixel value at (x={x}, y={y}): {pixel_val}")
```
