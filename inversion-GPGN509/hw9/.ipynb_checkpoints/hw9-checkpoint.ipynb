```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the file into a structured NumPy array
data = np.loadtxt("HW9data.txt", skiprows=1)  # skiprows=1 to ignore the header row

# Extract each column
x_in = data[:, 0]
y_in = data[:, 1]
i_in = data[:, 2]
s_in = data[:, 3]
```

# **1. Individual inversion**

## **Do Bayestian Inversion based ONLY on the light intensity mersurement**

Prior Joint PDF (data: i and model: r)

r: no information therefore we have uniform distribution

We do not need to rescale PDF functions right now, because the final Posterior PDF is normalized 

```python
# i and r variables for the marginal pdf
nx = 1001
i_x = np.linspace(0.05, 1., nx)
r_x = np.linspace(0., 12., nx)
i0 = i_in[0]
s0 = s_in[0]
g_prior_mar_ir = np.exp(-((i_x - i0)** 2) / (2 * s0 ** 2))
g_prior_ir = np.zeros((nx, nx))
g_prior_ir[:, :] = g_prior_mar_ir
g_prior_ir = np.fliplr(g_prior_ir)

plt.imshow(g_prior_ir.T, extent=(r_x.min(), r_x.max(), i_x.min(), i_x.max()), aspect='auto')
plt.xlabel("r_x")
plt.ylabel("i_x")
plt.title("Prior Joint PDF i-r")
plt.show()
```

Theoretical Data

```python
i_R = 1
r0 = 1
i_theo = r0/(r0+r_x)*i_R
```

Sofar, we have no idea with R, we still have uniform distrution of R

```python
# Define the exponential decay function to handle array input
def compute_decay_theo_pdf(i, i_T, k=20):
    """
    Calculate the probability density function for an array of observed intensities i,
    with the condition that f(i) = 0 for elements where i > i_T.
    
    Parameters:
    i : np.array
        Array of observed light intensities.
    i_T : float
        Theoretical light intensity in clear weather.
    k : float, optional
        Decay constant, defaults to 20 a.u.^-1.
    
    Returns:
    np.array
        Array of probability densities for each observed intensity in i.
    """
    # Apply the condition element-wise: return 0 where i > i_T, otherwise calculate exp(-k * |i - i_T|)
    return np.where(i > i_T, 0, np.exp(-k * np.abs(i - i_T)))
```

Create a theoretical joint PDF from given equation (again, I do not do normalization here):

```python
probability_densities = []
for i in range(nx):
    i_T = i_theo[i]
    probability_densities.append(compute_decay_theo_pdf(i_x, i_T, k=20))
probability_densities = np.array(probability_densities)
probability_densities = np.fliplr(probability_densities)
plt.imshow(probability_densities.T, aspect='auto', extent=(r_x.min(), r_x.max(), i_x.min(), i_x.max()))
plt.xlabel("r_x")
plt.ylabel("i_x")
plt.title("Thoretical Joint PDF i-r")
```

Posterior PDF

```python
g = g_prior_ir * probability_densities
plt.imshow(g.T, extent=(r_x.min(), r_x.max(), i_x.min(), i_x.max()), aspect='auto')
plt.xlabel("r_x")
plt.ylabel("i_x")
plt.title("Posterial Joint PDF i-r")
```

Here I do the normalization for marginal pdf of r_x

```python
integral_over_y = np.trapezoid(g, i_x, axis=1)
integ = np.trapezoid(integral_over_y, r_x)
# Do nomalization:
integral_over_y = integral_over_y/integ
plt.plot(r_x, integral_over_y)
plt.xlabel("r_x")
plt.title("Posterial Marginal PDF r")
```

## **Now I do for all data:**

```python
# List to store integral results for each data point
marginal_r_post = []

# Loop over each entry in i_in and s_in
for i0, s0 in zip(i_in, s_in):
    # Calculate the marginal prior for the current data point
    g_prior_mar_ir = np.exp(-((i_x - i0)** 2) / (2 * s0 ** 2))
    
    # Create the 2D prior by repeating the marginal prior
    g_prior_ir = np.zeros((nx, nx))
    g_prior_ir[:, :] = g_prior_mar_ir
    g_prior_ir = np.fliplr(g_prior_ir)

    # Compute the theoretical intensity i_theo for the current r_x values
    i_theo = r0 / (r0 + r_x) * i_R

    # Calculate probability densities for the current i_theo
    probability_densities = []
    for i in range(nx):
        i_T = i_theo[i]
        probability_densities.append(compute_decay_theo_pdf(i_x, i_T, k=20))
    probability_densities = np.array(probability_densities)

    # Flip probability densities matrix if needed
    probability_densities = np.fliplr(probability_densities)

    # Combine prior and likelihood to get the posterior
    g = g_prior_ir * probability_densities

    # Integrate over r_x (y-axis) for the current posterior
    integral_over_y = np.trapezoid(g, i_x, axis=1)

    integ = np.trapezoid(integral_over_y, r_x)

    # Do nomalization:
    integral_over_y = integral_over_y/integ
    # Append the result for this data point to the list
    marginal_r_post.append(integral_over_y)

# Convert the list of integrals to a NumPy array for easier manipulation
marginal_r_post = np.array(marginal_r_post)
```

We have the distrubition of distance, given the position of the beacon, we will have probability of the locations. The new distrubition would be the prior joint pdf of location with x and y with the given beacon, but it is posterior of the light intensity inversion

```python
# Define the grid of distances with specified range
grid_size = 200  # Resolution of the grid
x = np.linspace(-6, 6, grid_size)  # Range for x from -6 to 6
y = np.linspace(-4, 4, grid_size)  # Range for y from -4 to 4
X, Y = np.meshgrid(x, y)


# Assume `r` is the distance array used for interpolation (e.g., same for all integrals)
r = r_x  # Make sure `r_x` is defined in your code

# Initialize a list to store the density arrays for each PDF in `all_integrals`
density_arrays_1 = []

# Loop through each PDF in `all_integrals` and compute density arrays
for i in range(len(marginal_r_post)):
    x0, y0 = x_in[i], y_in[i]

    # Calculate the distance from each point in the grid to the central point
    R = np.sqrt((X - x0)**2 + (Y - y0)**2)
    pdf_r = marginal_r_post[i]
    # Interpolate the PDF values onto the distance grid R
    density = np.interp(R, r, pdf_r)
    
    # Append the density array to the list
    density_arrays_1.append(density)

# Optionally, convert the list to a 3D NumPy array for easier manipulation
density_arrays_1 = np.array(density_arrays_1)  # Shape will be (num_integrals, grid_size, grid_size)
```

```python
plt.imshow(density_arrays_1[8], extent=(-6,6, -4,4 ), origin='lower')
plt.scatter(x_in[8], y_in[8])
plt.title("Distribution of positions with data from beacon[8]");
```

```python
combined_density = np.prod(density_arrays_1, axis=0)
combined_density = combined_density.T
plt.imshow(combined_density.T, extent=(-6,6, -4,4 ), origin='lower')
plt.title("Distribution of positions with data from all beacons");
```

Compute the posterial marginal prob. density function, with normalization

```python
integ
```

```python
x_marginal_post = np.trapezoid(combined_density, y, axis=1)
integ = np.trapezoid(x_marginal_post, x)
# Do nomarlization
x_marginal_post = x_marginal_post/integ
plt.plot(x, x_marginal_post)
plt.title("Marginal PDF of x position")
```

```python
y_marginal_post = np.trapezoid(combined_density, x, axis=0)
integ = np.trapezoid(y_marginal_post, y)
# Do nomarlization
plt.plot(y, y_marginal_post)
plt.title("Marginal PDF of x position")
```

```python
# Compute the weighted mean (center)
centerx = np.sum(x * x_marginal_post) / np.sum(x_marginal_post)
# Compute the weighted standard deviation (dispersion)
dispersionx = np.sqrt(np.sum(x_marginal_post * (x - centerx)**2) / np.sum(x_marginal_post))

# Compute the weighted mean (center)
centery = np.sum(x * y_marginal_post) / np.sum(y_marginal_post)
# Compute the weighted standard deviation (dispersion)
dispersiony = np.sqrt(np.sum(y_marginal_post * (y - centery)**2) / np.sum(y_marginal_post))

print("Center (Mean):", centerx ,centery)
print("Dispersion (Standard Deviation):", dispersionx, dispersiony)
```

# **2 . Cummulative inversion**

Assume that we have the theoretical lines: "you are pretty sure that you did not deviate a lot from this path while drifting without accurate positioning, although you do not know how far off course you might be". The data is taken from the the dashline in the PDF

```python
# Points given
x1, y1 = -3.893048128342246, 3.7009345794392523
x2, y2 = 3.8930481283422456, -3.6688918558077437

# Calculate the slope (m) and y-intercept (b) of the line
m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

# Define the x range and calculate corresponding y values
x_vals = np.linspace(-6, 6, 200)
y_vals = m * x_vals + b
sigma = 1.5
```

```python
y_x = np.linspace(-4, 4, 200)

# Define the array of x values
y = np.linspace(-4, 4, 200)

# Define the Gaussian function
def gaussian(x, mu, sigma):
    """
    Compute the Gaussian function for a given array of x, mean (mu), and standard deviation (sigma).
    
    Parameters:
    x (np.array): Array of x-values.
    mu (float): Mean of the Gaussian.
    sigma (float): Standard deviation of the Gaussian.
    
    Returns:
    np.array: Gaussian function values at each x.
    """
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Define the array of mu values (means) and the sigma value
sigma = 1.5  # Standard deviation

# Create a 2D array to store the Gaussian functions
g_theoretical_pos = np.array([gaussian(y, mu, sigma) for mu in y_vals])

g_theoretical_pos=np.fliplr(g_theoretical_pos)

plt.imshow(g_theoretical_pos.T, extent=(-6,6, -4,4 ))
plt.title("Theoretical Distribution based on Assumption - Non-normalized");
```

using Boxcar distribution

```python

# Define the array of x values
y = np.linspace(-4, 4, 200)
y_vals = np.linspace(-4, 4, 200)  # Centers for each boxcar distribution

# Define the boxcar width
width = 1.5  # This replaces sigma for boxcar width

# Define the Boxcar function (replacing Gaussian)
def boxcar(x, mu, width):
    """
    Compute the Boxcar function for a given array of x, center (mu), and width.
    
    Parameters:
    x (np.array): Array of x-values.
    mu (float): Center of the Boxcar.
    width (float): Width of the Boxcar.
    
    Returns:
    np.array: Boxcar function values at each x.
    """
    return np.where((x >= mu - width / 2) & (x <= mu + width / 2), 1, 0)

# Create a 2D array to store the Boxcar functions
g_theoretical_pos = np.array([boxcar(y, mu, width) for mu in y_vals])

# Plot the Boxcar-based distribution
plt.imshow(g_theoretical_pos.T, extent=(-6, 6, -4, 4), aspect='auto')
plt.title("Theoretical Distribution based on Assumption - Boxcar")
plt.xlabel("x (corresponding to y_vals)")
plt.ylabel("y (corresponding to y)")
plt.colorbar(label="Intensity")
plt.show()
```

Here, I do the animation for the first queston: Individual inversion

```python
plt.imshow(density_arrays_1[8], extent=(-6,6, -4,4 ), origin='lower')
plt.scatter(x_in[8], y_in[8])
plt.title("Distribution of positions with data from beacon[8]");
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib inline

cul = np.ones_like(density_arrays_1[0])
cul_list = []
for i in range(len(density_arrays_1)):
    cul = cul*density_arrays_1[i]
    # Integral over x-axis first (sum over rows for each column)
    integral_over_x = np.trapezoid(cul, x, axis=0)
    # Then, integral over y-axis (sum over columns for the result from above)
    total_integral = np.trapezoid(integral_over_x, y)
    cul=cul/total_integral
    cul_list.append(cul)

fig, ax = plt.subplots()
extent = (-6, 6, -4, 4)
im = ax.imshow(cul_list[0], extent=extent, origin='lower', cmap='viridis')
plt.colorbar(im, ax=ax)
ax.set_title("Density Array at Frame 1")
ax.set_xlabel("X Coordinate")
ax.set_ylabel("Y Coordinate")

# Animation update function
def update(frame):
    # Update the data in the imshow plot to show the current frame's matrix
    im.set_data(cul_list[frame])
    ax.set_title(f"Density Array at Frame {frame+1}")

# Create the animation
ani = FuncAnimation(fig, update, frames=len(cul_list), interval=200, repeat=True)


ani.save("density_array_animation.gif", writer='imagemagick')
```
